```python
new_str = re.sub(r'[^\w\s]', '', example_string)
print(new_str) 
```

# 2. Title: Cleansing Exercise 2 - Adapt the above code to capitalise the first letter of the iris species, using regular expressions

## 2.1. Calculations 

```python
import re
import urllib.request
from string import capwords
```

```python
# https://stackoverflow.com/a/1393367
#List comprehensions - Extracted from lecture content 

url = r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris = [line.decode('utf-8').strip() for line in urllib.request.urlopen(url)]

#iris
```

```python
stripIris = re.compile(r'([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),Iris-([a-z]+)')
```

<b>2.1.1. Option 1 </b>

```python
# Not ideal approach as whole word is capatilized rather than first letter. Refer to refernce 6 
[stripIris.sub(r'\5,\4,\3,\2,\1', line) for line in iris if line]

irisdata = [stripIris.sub(r'\5,\4,\3,\2,\1', line) for line in iris if line]

output1 = [x.upper() for x in irisdata]

print(output1)
```

<b>2.1.2. Option 2 </b>

```python
#See reference 6
output2 = [item.capitalize() for item in irisdata]
output2
```

<b>2.1.3 Option 3 </b>

```python
#Refernce 5 
output3 = [line.title() for line in irisdata]
output3
```

<b>2.1.4 Option 4 </b>

```python
#See references 1-4

def capitalize(match):
    letter=match.group()
    if letter.isalpha():
        return letter.capitalize()
    
    
output4 = [re.sub(r'[a-z]+', capitalize, line) for line in irisdata if line]
output4
```
