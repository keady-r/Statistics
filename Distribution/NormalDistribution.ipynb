<b> 1.1. Defining Normal Distribution

Looking up normal distibution in the dictionary the description returned is 'a function that presents the distribution of many random variable as a symmetrical bell-shaped graph' [1.] Simply it is a symmetrical probability distribution with two key parameter; standard deviation and mean [2]. 
    

<b> Mean
    
In a normal distribution the Mean is in the center of the grpah i.e. the mean is the peak of the bell curve. 

<b> Standard Deviation
    
Standard deviation, the measure of variability, defines the width of normal distribution and how far from the mean values tend to fall. 
    


```python
#Normal Distribution using random data reference 6.
import numpy as np
import matplotlib.pyplot as plt

data = np.random.default_rng()
values = plt.hist(data.normal(2, 5, 10000),
                  bins=100, density=True)
plt.show()
```

```python
#See reference 4. 
from scipy.stats import norm
import statistics
  
# Gather Data point for xaxis and determine mean and standard deviation 
x = np.arange(-100, 100, 0.5)
mean = statistics.mean(x)
std = statistics.stdev(x)
  
plt.plot(x, norm.pdf(x, mean, std))
plt.show()
```

<b> Normal Distribution in Nature 
    
A Normal Distribution, also defined as a bell curve or Gaussian distribution, is one of the most common forms of distribution that appear in natural phenomonas. Characterisitics that arise due to many independent processes that commonly form a normal distribution examples include, height, IQ scores, bloodpressure [2,3]. For this reason normal distributions have been studied in great detail and may theories have been put forward on why these distributions are so common. One such theory is the central limit theorem (CLT). 
    
<b> Central Limit theory 
    
The Central Limit theory states that "the averages calculated from independent, identically distributed random variables have approximately normal distributions independent of the type of distribution from which the variables are sampled". (See calculation below for reference) 

```python
# Reference 8 - plotting with central limit theory

import numpy
import matplotlib.pyplot as plt
 
# values 
values = [1, 10, 50, 100] 
# list of sample means
means = [] 
 
# Generating 1, 10, 30, 100 random numbers from -400 to 400
# taking their mean and appending it to list means.
for value in values:
    # Generating seed so that we can get same result
    # every time the loop is run...
    numpy.random.seed(1)
    x = [numpy.mean(numpy.random.randint( -500, 500, value)) for _i in range(1000)]
    means.append(x)
k = 0
 
# plotting two row and two columns figure size 10. 
fig, ax = plt.subplots(2, 2, figsize =(10, 10))
for i in range(0, 2):
    for value in range(0, 2):
        # Histogram for each x stored in means
        ax[i, value].hist(means[k], 100, density = True)
        ax[i, value].set_title(label = values[k])
        k = k + 1
plt.show()
```


<b> Empirical rule 
    
Empirical rule describes the percentage of data that falls within specific numbers of standard deviations from the mean for bell-shaped curves.[2]. For normal distribution 68.2% of all obervations will arise within +/- one std of the mean while 95.4% will fall within +/- two std of the mean and 99.7% within three std. [3,5]
